```python
#methods for feature excration
def common_words(sd):
    l1 = set(map(lambda w: w.lower().strip(), sd['question1'].split(" ")))
    l2 = set(map(lambda w: w.lower().strip(), sd['question2'].split(" ")))    
    return 1.0 * len(l1 & l2)
def Total_words(sd):
    w1 = set(map(lambda word: word.lower().strip(), sd['question1'].split(" ")))
    w2 = set(map(lambda word: word.lower().strip(), sd['question2'].split(" ")))    
    return 1.0 * (len(w1) + len(w2))
def shared_words(sd):
    w1 = set(map(lambda word: word.lower().strip(), sd['question1'].split(" ")))
    w2 = set(map(lambda word: word.lower().strip(), sd['question2'].split(" ")))    
    return 1.0 * len(w1 & w2)/(len(w1) + len(w2))
```

```python
#basic feature extraction 
df['frequency_qid1'] = df.groupby('qid1')['qid1'].transform('count') 
df['frequency_qid2'] = df.groupby('qid2')['qid2'].transform('count')
df['Q1len'] = df['question1'].str.len() 
df['Q2len'] = df['question2'].str.len()
df['Q1_nwords'] = df['question1'].apply(lambda x: len(x.split(" ")))
df['Q2_nwords'] = df['question2'].apply(lambda x: len(x.split(" ")))
df['common_word'] = df.apply(common_words, axis=1)
df['Total_words'] = df.apply(Total_words, axis=1)
df['share_words'] = df.apply(shared_words, axis=1)
df['frequency_q1+q2'] = df['frequency_qid1']+df['frequency_qid2']
df['frequency_q1-q2'] = abs(df['frequency_qid1']-df['frequency_qid2'])

```
